```python
year = list(range(1960, 2016))
for i in year:    
    df6 = df3.unstack(1)
    df6 = df6[i]
    df6 = df6.unstack(1)
    s=df6['population']/1000000
    df6.plot.scatter('fertility', 'lifeexp', s=s)
    plt.savefig(f'myplot_{i}.png')
```

```python
import imageio

images = []

year = list(range(1960, 2016))

for i in year:
    filename = f'lifeexp_{i}.png'.format(i)
    images.append(imageio.imread(f'myplot_{i}.png'))

imageio.mimsave('output.gif', images, fps=20)
```
