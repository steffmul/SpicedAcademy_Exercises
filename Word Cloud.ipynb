```python
import numpy as np
from matplotlib import pyplot as plt
import wordcloud
```

```python
# lyrics by Zheng lab "Bad project"
lyrics = open('bad_project.txt').read()
```

```python
mask = np.zeros((500, 500, 3), np.uint8)

mask.shape

#mask[150:350,50:350,:] = 255  # masked out area

cloud = wordcloud.WordCloud(background_color="white",
                max_words=150,
                mask=mask,
                collocations=False,  # calculates frequencies
                contour_color='steelblue').generate(lyrics) # <-- lyrics is a string
                # stop words are removed!

plt.figure(figsize = (40,40))
plt.imshow(cloud, interpolation='bilinear')
plt.savefig('wordcloud.png')
plt.xticks([])
plt.yticks([])
plt.show()
```
